```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
# irisデータのロード
from sklearn.datasets import load_iris
iris = load_iris()
print(iris.DESCR)
```

```python
# [がく片の長さ(cm), がく片の幅(cm), 花びらの長さ(cm), 花びらの幅(cm)]
columns = iris.feature_names
columns
```

```python
X = iris.data
X[:5]
```

```python
y = iris.target
y
```

```python
X_df = pd.DataFrame(X, columns=columns)
y_df = pd.DataFrame(y, columns=['speacies'])
df = pd.concat([X_df, y_df], axis=1)
df.head()
```

```python
df.describe()
```

```python
import seaborn as sns
sns.pairplot(df, hue='speacies', size=2.0)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
X_std = scaler.fit_transform(X)
```

```python
X_1 = X_std[:100, [2,3]]
y_1 = y[:100]
```

```python
plt.scatter(X_std[:50, 2], X_std[:50, 3], color='red', marker='o', label='setosa')
plt.scatter(X_std[50:100, 2], X_std[50:100, 3], color='blue', marker='x', label='versicolor')
plt.ylabel('petal length [standarlized]')
plt.xlabel('petal width [standarlezed]')
plt.legend(loc='upper left')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_1, y_1, test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import Perceptron
ppn_1 = Perceptron(eta0=0.1)
ppn_1.fit(X_train, y_train)
```

```python
# ２次元データの決定曲線をプロットする関数
# (X, y, 分類器)
def plot_decision_regions(X, y, classifier, resolution=0.02):
    from matplotlib.colors import ListedColormap
    markers = ("s", "x", "o", "^", "v")
    colors = ("red", "blue", "lightgreen", "gray", "cyan")
    cmap = ListedColormap(colors[:len(np.unique(y))])
    x1_min, x1_max = X[:, 0].min()-1 , X[:, 0].max() +1
    x2_min, x2_max = X[:, 1].min()-1 , X[:, 1].max() +1
    
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution), np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.4, cmap=cmap)
    
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())
    
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], y=X[y == cl, 1], alpha=0.8, c=cmap(idx),marker=markers[idx], label=cl)
```

```python
# 散布図上に決定曲線を出力
plot_decision_regions(X_1, y_1, ppn_1)
plt.xlabel("petal length [standardized]")
plt.ylabel("petal width [standardized]")
plt.legend(loc="upper left")
plt.title('Setosa or Versicolour')
plt.show()
```

```python
X_2 = X_std[:, [2,3]]
y_2 = y
X_2_train, X_2_test, y_2_train, y_2_test = train_test_split(X_2, y_2, test_size=0.3, random_state=0)
```

```python
ppn_2 = Perceptron(eta0=0.1)
ppn_2.fit(X_2_train, y_2_train)
```

```python
# 散布図上に決定曲線を出力
plot_decision_regions(X_2, y_2, ppn_2)
plt.xlabel("petal length [standardized]")
plt.ylabel("petal width [standardized]")
plt.legend(loc="upper left")
plt.title('Setosa or Versicolour')
plt.show()
```

```python
print(ppn_2.score(X_2_train, y_2_train))
print(ppn_2.score(X_2_test, y_2_test))
```
